# Part 1

### Get the data ready

```python
# Libraries for scrapping the wikipedia website

from bs4 import BeautifulSoup
import html5lib
```

```python
# Collect the data from the wikipedia

wiki_url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

result = requests.get(wiki_url)
```

```python
# Create the BeautifulSOup object and get the content of scrapped data

bs = BeautifulSoup(result.content,'html5lib')
bs
```

```python
# Find the table tag

table = bs.find('table')
table
```

```python
# Create a new dataframe where scrapped data will be stored

df = pd.DataFrame(columns=['Postal Code','Borough'])
df
```

```python
# Now scrap the table to exract the data

code = []
borough = []

for rows in table.findAll('tr'):
    
    for td in rows.findAll('td'):
        
        #code.append(td.p.text[:3])
        code.append(td.p.text[:3])
        borough.append(td.p.text[3:])


df['Postal Code'] = code

df['Borough'] = borough
```

```python
df.head()
```

```python
### Create a new columns 'Neighborhood' and we will extract the neighborhood names from the 'Borough' column i.e the data in brackets

df['Neighborhood'] = df['Borough'].str.extract('(\(.*\))',expand=False)

df.head()
```

```python
### As we created a Neighborhood column and extracted the neighborhood names, now we will remove that data from the 'Borough' column 

df['Borough'] = df['Borough'].str.replace('(\(.*\))','',regex=True)

df.head()
```

```python
### Now we can see that there is escape sequence characters in the Borough column i.e '\n'. Lets remove that

df['Borough'] = df['Borough'].str.replace('\\n','',regex=True)

df.head()
```

```python
### Now remove the brackets and , from the Neighborhood column

df['Neighborhood'] = df['Neighborhood'].str.replace('[\(\)]','',regex=True)

df['Neighborhood'] = df['Neighborhood'].str.replace('/',' , ')

df['Neighborhood'] = df['Neighborhood'].str.replace(' , ',', ')

df['Neighborhood'] = df['Neighborhood'].str.replace(' , ',', ')

df
```

```python
### Lets check the total counts on the basis of Borough columns

df['Borough'].value_counts()
```

#### As we can see that there are 77 Boroughs which are not assigned, lets get the more details for those rows 

```python
# Create a new dataframe and store those 77 rows in this dataframe

new_df = df.copy()

filtered = df[df['Borough']=='Not assigned' ]

filtered = filtered.reset_index(drop=True)

filtered
```

```python
# Lets check if these 'Not assigned Borough names' contains 'Not assigned' Neighborhood names, if yes then drop all those rows

filtered['Neighborhood'] = filtered['Neighborhood'].replace(np.nan,'Not assigned')

filtered
```

```python
filtered.dtypes
```

```python
filtered['Neighborhood'].value_counts()
```

#### So we have 77 Boroughs name with 'Not assigned' names and 'Not assigned' Neighborhood names, lets drop all 77 rows from the original dataframe

```python
# Lets drop the rows with 'Not assigned' values

df = df.dropna(axis=0)

df = df.reset_index(drop=True)

df
```

#### So we got 103 observations and 3 attributes from the original dataframe

### Create a function that will extract the categories of these venues

```python
# Function that helps us to get the venues of the Roselawn Neighborhood

def getCategoriesData(df):
    
    try:
        
        cat_lst = df['venue']
        
    except:
        
        cat_lst = df['venue.categories']
        
    if len(cat_lst)==0:
        
        return None
    
    else:
        
        return cat_lst[0]['name']
```

```python
venues_df_filtered['venue.categories'] = venues_df_filtered.apply(getCategoriesData,axis=1)

venues_df_filtered
```

### Now lets explore the the neighborhoods in Toronto as we did before for one neighborhood 'Roselwan'

```python
def getVenuesDetails(names,latitudes,longitudes):
    
    venues_data = []
    
    for name,lat,lng in zip(names,latitudes,longitudes):
        
        url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(cid,csec,v,lat,lng,radius,lim)
        
        result = requests.get(url).json()
        
        results = result['response']['groups'][0]['items']
        
        venues_data.append([(name,
                             lat,
                             lng,
                             venue['venue']['name'],
                             venue['venue']['location']['lat'],
                             venue['venue']['location']['lng'],
                             venue['venue']['categories'][0]['name'],
                            ) for venue in results])
        
    venues_result = pd.DataFrame([venue for venues in venues_data for venue in venues])
    
    venues_result.columns = ['Neighborhood',
                             'Latitude',
                             'Longitude',
                             'Venue Name',
                             'Venue Latitude',
                             'Venue Longitude',
                             'Venue Category'                            
                            ]
    
    return (venues_result)
```

```python
venues_df = getVenuesDetails(names=df['Neighborhood'],latitudes=df['Latitude'],longitudes=df['Longitude'])

venues_df
```

### Lets create a new dataframe which contains the Venue Categories in a one hot encoded form

```python
venue_cat_encoded = pd.get_dummies(venues_df[['Venue Category']])

venue_cat_encoded
```

### As we can see the Neighborhood column is added at the last column index, lets bring it to first position of column index

```python
columns = [venue_cat_encoded.columns[-1]] + list(venue_cat_encoded.columns[:-1])

venue_cat_encoded = venue_cat_encoded[columns]

venue_cat_encoded
```

### Lets add the Neighborhood columns in encoded data

```python
venue_cat_encoded['Neighborhood'] = venues_df['Neighborhood']

venue_cat_encoded
```

### Now lets Create a new dataframe which consists of frequency of occurence of every venue in a Neighborhood

```python
venue_encoded_mean = venue_cat_encoded.groupby('Neighborhood').mean().reset_index()

venue_encoded_mean
```

```python
venue_encoded_mean.shape
```

#### So we have 39 observations of 237 attributes

### Now lets find the top 10 venues from the encoded data

```python
def getTopVenues(data,top_num):
    
    d = data.iloc[1:]
    
    sorted_data = d.sort_values(ascending=False)
    
    return sorted_data.index.values[0:top_num]
```

```python
top_num = 10

positions = ['st','nd','rd']

columns = ['Neighborhood']

for pos in np.arange(top_num):
    
    try:
        
        columns.append('{}{} common venue'.format(pos+1,positions[pos]))
        
    except:
        
        columns.append('{}th common venue'.format(pos+1))
    
common_venues_df = pd.DataFrame(columns=columns)

common_venues_df['Neighborhood'] = venue_encoded_mean['Neighborhood']

for pos in np.arange(venue_encoded_mean.shape[0]):
    
    common_venues_df.iloc[pos,1:] = getTopVenues(venue_encoded_mean.iloc[pos,:],top_num)
```

## Data Clustering

```python
venue_encoded_mean.head()
```

```python
clusters = 6

new_df = venue_encoded_mean.drop('Neighborhood',axis=1)

kmeans = KMeans(n_clusters=clusters,random_state=0)

kmeans.fit(new_df)
```

```python
labels = kmeans.labels_
labels
```
